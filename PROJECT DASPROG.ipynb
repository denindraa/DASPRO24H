```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('upah.csv')

plt.figure(figsize=(14,6))
plt.bar(df['provinsi'], df['upah.2022'], color='skyblue')
plt.title('Upah Miinimum provinsi di Tahun 2022')
plt.xlabel('Provinsi')
plt.ylabel('Upah Minimum(Ribu Rupiah)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('upah.csv')
tahun = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

plt.figure(figsize=(14,8))
for i in range(len(df)):
    provinsi = df.loc[i, 'provinsi']
    upah = df.iloc[i, 1:].values 
    plt.plot(tahun, upah, marker='o', label=provinsi)
plt.title('Trend Upah Minimum Semua Provinsi (20152022)')
plt.xlabel('Tahun')
plt.ylabel('Upah Minimum (Ribu Rupiah)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('upah.csv')
plt.figure(figsize=(15,15))
plt.pie(df['upah.2022'], labels=df['provinsi'], autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Upah Minimum Antar Provinsi Tahun 2022')
plt.axis('equal')  
plt.tight_layout()
plt.show()
```
